```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv('temp_data.csv', parse_dates=True, index_col= 'Date Time')
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df[['Tout','GasUsed']]['2018-11-1':'2018-11-22'].values
y = df['Tin']['2018-11-1':'2018-11-22'].values
```

```python
X_test = df[['Tout','GasUsed']]['2018-11-23':'2018-11-30'].values
y_test = df['Tin']['2018-11-23':'2018-11-30'].values
```

```python
lm = LinearRegression()
regressor = lm.fit(X, y)
```

```python
y_pred = regressor.predict(X_test)
```

```python
np.mean((y_pred-y_test)**2), np.max(abs(y_pred-y_test))
```

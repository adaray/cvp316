```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('measured_values.csv', skiprows=1, parse_dates=True, index_col=0)
```

```python
temp = df[['ROOM1']]
temp.columns = ['Temp']
```

```python
temp.head()
```

```python
temp.plot()
```

```python
#HP filter
result = sm.tsa.filters.hpfilter(temp['Temp'])
```

```python
result[0].plot()
```

```python
temp['Temp'].plot()
result[1].plot()
```

```python
#Moving average - rolling
temp['6HR'] = temp['Temp'].rolling(window=6).mean()
```

```python
temp[['Temp','6HR']].plot()
```

```python
#Exponential Weighted Moving Average
temp['EWMA-6HR'] = temp['Temp'].ewm(span=6).mean()
```

```python
temp[['6HR','EWMA-6HR']].plot()
```

```python
#Rolling mean and standard deviation
temp['Temp'].rolling(6).mean().plot()
temp['Temp'].rolling(6).std().plot()
```

```python
#ETS decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result = seasonal_decompose(temp['Temp'])
```

```python
fig = result.plot()
```

```python
#ARIMA code
from statsmodels.tsa.stattools import adfuller
```

```python
def adcheck(x):
    result = adfuller(x)
    print('Dicky Fuller Test')
    lab = ['ADF Test Statistic','p-value','# of lags','No. of Obsr.']
    for value, label in zip(result,lab):
        print(label,':',str(value))
    if result[1] <= 0.05:
        print('Is stationary')
    else:
        print('Not stationary')
```

```python
adcheck(temp['Temp'])
```

```python
temp['TempDiff1'] = temp['Temp'] - temp['Temp'].shift(1)
```

```python
temp['TempDiff1'].dropna(inplace=True)
```

```python
adcheck(temp['TempDiff1'])
```

```python
temp['TempDiff1'].plot()
```

```python
#PROPER ARIMA MODEL
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = sm.tsa.statespace.SARIMAX(temp['Temp'], order=(0,1,0), seasonal_order=(1,1,1,1))
```

```python
result = model.fit()
```

```python
result.resid.plot()
```

```python
temp['forecast'] = result.predict(start='2018-11-28', end='2018-11-30')
```

```python
temp[['Temp','forecast']].plot()
```

```python
temp['forecast'].dropna(inplace=True)
```

```python
max(abs((temp['forecast'] - temp['Temp']).dropna()))
```
